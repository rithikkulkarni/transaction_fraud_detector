Import Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection/Processing

```python
fraud_data = pd.read_csv('/content/fraud_detection_quantitative_dataset.csv', header=None)
```

```python
fraud_data.head()
```

```python
fraud_data.shape
```

```python
fraud_data.describe()
```

```python
fraud_data[16].value_counts()
```

```python
fraud_data.groupby(50).mean()
```

```python
print(fraud_data)
```

```python
# Separating the data and labels
X = fraud_data.drop(columns=50, axis=1)
Y = fraud_data[50]
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

```python
model = LogisticRegression()
```

```python
# Now we train the Logistic Regression model using our training data
model.fit(X_train, Y_train)
```

```python
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data: ', training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data: ', test_data_accuracy)
```
